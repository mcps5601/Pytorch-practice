```python
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torchvision.utils import make_grid
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
from PIL import Image
```

```python
input_size=128
batch_size=16
```

```python
TRAIN_DATA_PATH = '/data/home/dean/imagebank/PXR/pelvic_fracture_all/'
```

```python
transform = transforms.Compose([transforms.Resize((input_size, input_size)), transforms.ToTensor(), transforms.Normalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5))])
train_data = datasets.ImageFolder(root=TRAIN_DATA_PATH, transform=transform)
data_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
```

```python
# iterations of dataloader
dataiter = iter(data_loader)
```

```python
# test loading data with cv2
hip_img_list=[]
fracture_img_list=[]
normal_img_list=[]

for root, dirs, files in os.walk(TRAIN_DATA_PATH):
    for file in files:
        filename = os.path.join(root, file)
        print (filename)
        category_name = os.path.basename(root)
        im = cv2.imread(filename)
        im = cv2.resize(im, (input_size, input_size), interpolation=cv2.INTER_LINEAR)
        imarray = np.array(im)
        if category_name == 'hip':
            hip_img_list.append(imarray)
        if category_name == 'fracture':
            fracture_img_list.append(imarray)
        if category_name == 'normal':
            normal_img_list.append(imarray)
```
